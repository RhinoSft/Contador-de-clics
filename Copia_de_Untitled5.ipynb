```python
barrios = ['Ate', 'Barranco', 'Comas', 'Lince', 'El Agustino', 'San Luis', 'Callao']

seleccion = datos['Distrito'].isin(barrios)
datos = datos[seleccion]
```

```python
datos['Distrito'].drop_duplicates()
```

```python
grupo_barrio = datos.groupby('Distrito')
```

```python
grupo_barrio.groups
```

```python
for barrio, data in grupo_barrio:
  print('{}->{}'.format(barrio, data.Valor.mean().round(2)))
```

```python
grupo_barrio[['Valor', 'Mantenimiento', 'Impuesto']].mean().round(2)
```

```python
grupo_barrio['Valor'].describe().round(2)
```

```python
grupo_barrio['Valor'].aggregate(['min', 'max',]).rename(columns = {'min': 'Minimo', 'max': 'Maximo',})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_barrio['Valor'].mean().plot.bar(color = 'green')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor promedio del Alquiler por distrito', {'fontsize': 22})
```

```python
fig = grupo_barrio['Valor'].max().plot.line(color = 'blue')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor maximo por Distrito', {'fontsize': 22})
```
